```python
import pandas as pd
import seaborn as sns
import numpy as np
sns.set_theme(style="whitegrid")
```

```python
sns.__version__
```

```python
amex_logs = pd.read_csv("AMEX Users logs.csv")
```

```python
amex_logs
```

```python
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.str.split.html
amex_logs[['Location','Location2']] = amex_logs.activity.str.split(pat="-", expand = True)
```

```python
amex_logs['tempC'] = (amex_logs['temperature']-32)*5/9
```

```python
amex_logs['hsiC'] = (amex_logs['hsi']-32)*5/9
```

```python
amex_logs
```

```python
amex_logs['Location'].value_counts().sort_index()
```

```python
amex_logs['Location2'].value_counts().sort_index()
```

```python
amex_logs['Location2']= amex_logs['Location2'].str.replace('Restaurants','Restaurant')
```

```python
amex_logs['Location2']= amex_logs['Location2'].str.replace('notes','Other')
```

```python
amex_logs['Location2'].value_counts()
```

```python
amex_logs_hsi = amex_logs[amex_logs.hsi != 0]
```

```python
amex_logs_hsi.describe()
```

```python
amex_logs
```

```python
amex_logs = amex_logs.sort_values(by=['hsiC'])
```

```python
ax = sns.boxplot(x="Location2", y="hsiC", hue="Location",data=amex_logs_hsi, palette="Set2", linewidth=1)
```

```python
# https://stackoverflow.com/questions/16183462/saving-images-in-python-at-a-very-high-quality
# https://stackoverflow.com/questions/49930064/get-sample-size-for-boxplots-in-seaborn-factorplot
import matplotlib.pyplot as plt
# figsize set up
fig, ax = plt.subplots(figsize=(12,8))
ax = sns.boxplot(x="Location2", y="hsiC", hue="Location", data=amex_logs_hsi, palette="Set3", linewidth=1)
plt.ylabel("Heat Stress Index (Celsius)", size=14)
plt.xlabel("Location", size=14)
plt.legend( loc='upper left')

medians = amex_logs_hsi.groupby(['Location2','Location'])['hsiC'].median()
nobs =  amex_logs_hsi.groupby(['Location2','Location']).apply(lambda x: 'n: {}'.format(len(x)))

for ax in plt.gcf().axes:

    for tick, label in enumerate(ax.get_xticklabels()):
        ax_Location2 = label.get_text()

        for j, ax_Location in enumerate(ax.get_legend_handles_labels()[1]):
            x_offset = (j - 0.5) * 2/5
            med_val = medians[ax_Location2, ax_Location]
            num = nobs[ax_Location2, ax_Location]

            ax.text(tick + x_offset, med_val + 0.1, num,
                    horizontalalignment='center', size='x-small', color='red', weight='semibold')


#fig.savefig('Figure9.tif', format='tif', dpi=300)
```

```python
amex_logs["count"] = 1
```

```python
amex_logs
```

```python
# group by categorical variables
# https://stackoverflow.com/questions/31080383/counting-categorical-data-pandas-group-by-dataframe
amex_logs_R1 = amex_logs.groupby(['Location', 'Location2'])['count'].sum().reset_index()
```

```python
amex_logs_R1
```

```python
amex_logs_R1['percentage'] = 100*amex_logs_R1['count']/sum(amex_logs_R1['count'])
```

```python
amex_logs_R1['percentage'] = amex_logs_R1['percentage'].round(2)
```

```python
amex_logs_R1 = amex_logs_R1.sort_values(by=['percentage'])
```

```python
# reverse the lines
amex_logs_R2 = amex_logs_R1.iloc[::-1]
```

```python
amex_logs_R2
```

```python
# Draw the grouped_barplot
# https://seaborn.pydata.org/examples/grouped_barplot.html
# https://stackoverflow.com/questions/53029604/how-to-add-bar-value-on-the-top-catplot-seaborn

g = sns.catplot(
    data=amex_logs_R2, kind="bar",
    x="Location2", y="percentage", hue="Location",ci="sd", palette="Set1", height=6, legend = False)

g.fig.set_size_inches(12, 8)
g.despine(left=True)
g.set_axis_labels("Location", "Percentage(%)", size= 14)
plt.legend(loc='upper right')

# extract the matplotlib axes_subplot objects from the FacetGrid
ax = g.facet_axis(0, 0)
 
for p in ax.patches:
    ax.text(p.get_x() + 0.02, 
            p.get_height() * 1.02, 
            '{0:.2f}%'.format(p.get_height()), 
            color='black', rotation='horizontal', size='large')
g.savefig('Figure8.tif', format='tif', dpi=300)
```
