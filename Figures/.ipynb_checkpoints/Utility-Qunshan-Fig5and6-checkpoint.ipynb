```python
Location = ['Outdoor','Indoor',]
ua_logs_hsi_R1 = ua_logs_hsi_final[ua_logs_hsi_final.LOCATION.isin(Location)] 
```

```python
ua_logs_hsi_R1
```

```python
import matplotlib.pyplot as plt
# figsize set up
fig, ax = plt.subplots(figsize=(12,8))
# hue order: https://seaborn.pydata.org/generated/seaborn.boxplot.html
# Always check the original function
ax = sns.boxplot(x="SPECIFIC LOCATION", y="Heat Stress Index", hue="LOCATION", hue_order = ["Outdoor","Indoor"], data=ua_logs_hsi_R1, order=["Home", "Exercise", "Work", "Shopping", "Transit", "Other","Restaurant"], palette="Set3", linewidth=1)
plt.ylabel("Heat Stress Index (Celsius)", size=14)
plt.xlabel("Location", size=14)
plt.legend( loc='upper left')
# change legend order
# https://github.com/mwaskom/seaborn/issues/720
#handles, _ = ax.get_legend_handles_labels()
#ax.legend(handles, ["Outdoor", "Indoor"])

medians = ua_logs_hsi_R1.groupby(['SPECIFIC LOCATION','LOCATION'])['Heat Stress Index'].median()
nobs =  ua_logs_hsi_R1.groupby(['SPECIFIC LOCATION','LOCATION']).apply(lambda x: 'n: {}'.format(len(x)))

for ax in plt.gcf().axes:

    for tick, label in enumerate(ax.get_xticklabels()):
        ax_Location2 = label.get_text()

        for j, ax_Location in enumerate(ax.get_legend_handles_labels()[1]):
            x_offset = (j - 0.5) * 2/5
            med_val = medians[ax_Location2, ax_Location]
            num = nobs[ax_Location2, ax_Location]

            ax.text(tick + x_offset, med_val + 0.1, num,
                    horizontalalignment='center', size='x-small', color='red', weight='semibold')


#fig.savefig('Figure6.tif', format='tif', dpi=300)
```

```python
ua_logs_hsi_R1["count"] = 1
```

```python
ua_logs_hsi_R1
```

```python
# group by categorical variables
# https://stackoverflow.com/questions/31080383/counting-categorical-data-pandas-group-by-dataframe
ua_logs_hsi_R2 = ua_logs_hsi_R1.groupby(['SPECIFIC LOCATION','LOCATION'])['count'].sum().reset_index()
```

```python
ua_logs_hsi_R2
```

```python
ua_logs_hsi_R2['percentage'] = (100*ua_logs_hsi_R2['count']/sum(ua_logs_hsi_R2['count'])).round(2)
```

```python
ua_logs_hsi_R2 = ua_logs_hsi_R2.sort_values(by=['percentage'])
```

```python
# reverse the lines
ua_logs_hsi_R3 = ua_logs_hsi_R2.iloc[::-1]
```

```python
ua_logs_hsi_R3
```

```python
# Draw the grouped_barplot
# https://seaborn.pydata.org/examples/grouped_barplot.html
# https://stackoverflow.com/questions/53029604/how-to-add-bar-value-on-the-top-catplot-seaborn

g = sns.catplot(
    data=ua_logs_hsi_R3, kind="bar",
    x="SPECIFIC LOCATION", y="percentage", hue="LOCATION",ci="sd", palette="Set1", height=6, legend = False)

g.fig.set_size_inches(12, 8)
g.despine(left=True)
g.set_axis_labels("LOCATION", "Percentage(%)", size= 14)
plt.legend(loc='upper right')

# extract the matplotlib axes_subplot objects from the FacetGrid
ax = g.facet_axis(0, 0)
 
for p in ax.patches:
    ax.text(p.get_x() + 0.02, 
            p.get_height() * 1.02, 
            '{0:.2f}%'.format(p.get_height()), 
            color='black', rotation='horizontal', size='large')
g.savefig('Figure5.tif', format='tif', dpi=300)
```
